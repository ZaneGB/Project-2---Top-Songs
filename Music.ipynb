```python
# Combine the data into a single dataset
music_data_complete = pd.merge(rank_data, esoteric_data, how="left", left_on=["Song"], right_on=["title"])  

#music_data_complete.head()
music_data_complete
```

```python
music_data_complete.drop(columns=["title"], inplace=True)
music_data_complete.head()
```

```python
music_data_complete = music_data_complete.rename(columns={"duration": "Duration", "familiarity":
"Familiarity", "location": "Location", "loudness": "Loudness", "tempo": "Tempo", "terms": "Genre"})

music_data_complete.head()

```

```python
music_data_complete= music_data_complete[["Rank", "Song", "Artist", 
"Year", "Genre", "Duration", "Location", "Familiarity",
"Loudness", "Tempo", "Score"]]

music_data_complete.head()
```

```python
music_data_complete.drop_duplicates(subset="Song", keep='first', inplace=True)
#music_data_complete
music_data_complete.head()
```

```python

music_data_complete["Duration"] = music_data_complete["Duration"].round(1)
music_data_complete["Loudness"] = (music_data_complete["Loudness"] * -1).round(1)
music_data_complete["Tempo"] = music_data_complete["Tempo"].round(1)
music_data_complete["Familiarity"] = (music_data_complete["Familiarity"] * 100).round(3)
music_data_complete.head()
```

```python
# music_data_complete["ArtistSongCount"] =np.nan
# music_data_complete["ArtistRank"] =np.nan
```

```python
artist_song_count = music_data_complete.groupby(["Artist"]).count()

artist_song_count.head()
```

```python
artist_song_count.drop(columns=["Rank", "Year","Genre", "Duration","Location","Familiarity",
"Loudness","Tempo","Score"], inplace=True)
artist_song_count = artist_song_count.rename(columns={"Song": "ArtistSongCount"})
artist_song_count.head()
```

```python
artist_song_count_db = artist_song_count.sort_values(["ArtistSongCount"], ascending=False)
artist_song_count_db.head()
```

```python
artist_song_count_db["ArtistRank"]=np.nan
#artist_song_count_db.head()
```

```python
artist_song_count_db = artist_song_count_db.reset_index(drop=False)
artist_song_count_db.head(10)
```

```python
new_rank = 1
row_count = 0
old_artist_song_count = artist_song_count_db["ArtistSongCount"].iloc[0]

#print(old_artist_song_count)

for index, row in artist_song_count_db.iterrows():
    row_count += 1
    if old_artist_song_count == row["ArtistSongCount"]:
        artist_song_count_db["ArtistRank"][index] = new_rank
    else:
        new_rank = row_count
        artist_song_count_db["ArtistRank"][index] = new_rank
        old_artist_song_count = row["ArtistSongCount"] 

artist_song_count_db.head(20)
```

```python
# Combine the data into a single dataset
music_db = pd.merge(music_data_complete, artist_song_count_db, how="left", left_on=["Artist"], right_on=["Artist"]) 

music_db.head(10)
```

```python
# Compute # of songs for each decade

fifties_count = 0
sixties_count = 0
seventies_count = 0
eighties_count = 0
nineties_count = 0
millennial_count = 0


for index, row in music_db.iterrows():
    
    if music_db["Year"][index] < 1960:
        fifties_count += 1    
    elif music_db["Year"][index] < 1970:
        sixties_count += 1
    elif music_db["Year"][index] < 1980:
        seventies_count += 1
    elif music_db["Year"][index] < 1990:
        eighties_count += 1
    elif music_db["Year"][index] < 2000:
        nineties_count += 1
    elif music_db["Year"][index] >= 2000:
        millennial_count += 1
        
print("50s Count: " + str(fifties_count))
print("60s_count: " + str(sixties_count))
print("70s_count: " + str(seventies_count))
print("80s_count: " + str(eighties_count))
print("90s_count: " + str(nineties_count))
print("2000s_count " + str(millennial_count))
```

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.orm import Session, relationship
from sqlalchemy import create_engine, inspect, func
# Import modules to declare columns and column data types
from sqlalchemy import Column, Integer, String, Float, ForeignKey
# Import and establish Base for which classes will be constructed 
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.sql import select
Base = declarative_base()
```

```python
#Create and connect to the sqlite database
engine = create_engine("sqlite:///music.sqlite", echo=False)
conn = engine.connect()
```

```python
# Create Song, Ranking and Artist classes
# ----------------------------------
class Performer(Base):
    __tablename__ = 'performer'
    id = Column(Integer, primary_key=True)
    artist = Column(String(255))
    artistsongcount = Column(Integer)
    artistrank = Column(Float)
    location = Column(String(255))
```

```python
# Create Track, Ranking and Artist classes
# ----------------------------------
class Track(Base):
    __tablename__ = 'track'
    id = Column(Integer, primary_key=True)
    song = Column(String(255))
    familiarity = Column(Float)
    duration = Column(Float)
    loudness = Column(Float)
    tempo = Column(Float)
    
    performer_id = Column(Integer, ForeignKey("performer.id"))
    performer_artist = relationship("artist", foreign_keys=[performer_id])
    
```

```python
# Create Track, Ranking and Artist classes
# ----------------------------------
class Ranking(Base):
    __tablename__ = 'ranking'
    id = Column(Integer, primary_key=True)
    rank = Column(Integer)
    song = Column(String(255))
    year = Column(Integer)
    genre = Column(String(255))
    score = Column(Float)
    
    track_id = Column(Integer, ForeignKey("track.id"))
    track_song = relationship("song", foreign_keys=[track_id])
```

```python
# Create both the table calsses within the database

Base.metadata.create_all(conn)
```
